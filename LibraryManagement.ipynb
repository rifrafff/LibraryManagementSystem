```python
class Library:
    def __init__(self, filename="books.txt"):
        self.filename = filename
        self.file = open(self.filename, 'a+')
        self.file.seek(0)

    def __del__(self):
        if hasattr(self, 'file') and self.file is not None:
              self.file.close()

    def list_books(self):
        self.file.seek(0)
        content = self.file.read()
        books_list = content.splitlines()

        if not books_list:
          print("There Is No Book")
        else:
          print("Books:")
          for book_info in books_list:
                  book_details = book_info.split(',')
                  book_name = book_details[0]
                  author = book_details[1]
                  print(f"Book Name: {book_name}, Author: {author}")

    def add_book(self):
        while True:
          book_name = input("Book Name: ")
          if book_name != "":
            break
          print("Cannot Be Left Blank")
        while True:
          author = input("Author: ")
          if author != "":
            break
          print("Cannot Be Left Blank")
        while True:
          try:
            release_year = input("Release Year: ")
            check = int(release_year)
            break
          except ValueError:
            print("Please Enter Number")
        while True:
          try:
            num_pages = input("Number of pages: ")
            check = int(num_pages)
            break
          except ValueError:
            print("Please Enter Number")

        book_info = f"{book_name},{author},{release_year},{num_pages}\n"
        self.file.write(book_info)
        print(book_name + " Book Added")

    def remove_book(self):
        content = self.file.read()
        books_list = content.splitlines()
        while True:
          find_book = input("Type The Name Of The Book You Want To Remove: ")
          if find_book != "":
            break
          print("Cannot Be Left Blank")

        if not books_list:
          print("There Is No Book")
        index = -1
        for i, book_info in enumerate(books_list):
            book_details = book_info.split(',')
            book_name = book_details[0]
            if find_book == book_name:
                index = i
                break

        if index != -1:
            del books_list[index]

            with open('books.txt', 'w') as file:
                file.truncate(0)

            with open('books.txt', 'a') as file:
                for book_info in books_list:
                    file.write(book_info + '\n')

            print(f"The Book '{find_book}' Has Been Removed.")
        else:
            print(f"The Book '{find_book}' Was Not Found.")

    def clear_list(self):
        self.file.seek(0)
        content = self.file.read()
        if not content:
          print("There Is No Book")
        else:
          with open('books.txt', 'w') as file:
              file.truncate(0)
          print("All Books Removed")

    def suggest_book(self):
        import random
        content = self.file.read()
        if not content:
          print("There Is No Book")
        else:
          books_list = content.splitlines()
          book_names = []
          for book_details in books_list:
            book_find = book_details.split(",")
            book_names.append(book_find[0])
          suggest = random.choice(book_names)
          print("You Can Read The ",suggest," Book")



while True:
  lib = Library()
  print("**********MENU**********\n","Type 1 To See The Book List\n","Type 2 To Add A Book\n","Type 3 To Remove A Book\n","Type 4 To Clear All Book List\n","Type 5 To Get Book Suggestion\n","Press q To Exit")
  action = input("Select Action : ")

  if action == "1":
    lib.list_books()
    del lib
  elif action == "2":
    lib.add_book()
    del lib
  elif action == "3":
    lib.remove_book()
    del lib
  elif action == "4":
    lib.clear_list()
    del lib
  elif action == "5":
    lib.suggest_book()
    del lib
  elif action == "q":
    break
  else:
    print("Invalid Action")
```
